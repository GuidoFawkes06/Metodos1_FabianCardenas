4. Calcular la probabilidad que n personas (n ≤ 365) tengan fechas diferentes de cumpleaños, i.e, escribir la formula general de cálculo. Grafique la probabilidad P(n ≤ 80) como funcion de n.

Hay 365 días en un año (ignorando los años bisiestos por simplicidad). Para la primera persona, hay 365 cumpleaños posibles, y para la segunda persona, hay 364 días restantes para tener un cumpleaños diferente, y así sucesivamente.Para encontrar la probabilidad de que n personas tengan cumpleaños diferentes, podemos usar la siguiente fórmula:
$$
p(n)=\frac{365}{365}*\frac{364}{365}*\frac{363}{365}*...*\frac{365-n}{365}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Probabilidad(n):
    return (np.math.factorial(int(365))/np.math.factorial(int(365)-n))*(int(365)**-n)
```

```python
N_=np.linspace(1,80,80)
p=np.zeros_like(N_)
for i in range(len(N_)):
    p[i]=Probabilidad(i)   
plt.scatter(N_,p)
```
